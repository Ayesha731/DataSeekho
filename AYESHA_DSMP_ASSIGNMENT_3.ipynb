# **Question#1**

class creation


```python
#create the class of book
class Book:
  #attributes of the books
  def __init__(self,title,author,pages):
    self.title=title
    self.author=author
    self.pages=pages
  #property of the book --method that display the title,author and pages of the book
  def get_info(self):
    print(f"The title of the book is {self.title}")
    print(f"The author of the book is {self.author}")
    print(f"The number of pages in the book is {self.pages}")
    print(f"The {self.title} is written by {self.author} and has total {self.pages}.\n")
#create the instance of object
book1=Book("Maseel","Noor Rajpoot",1400);
book2=Book("Namal","Nemrah Ahmad",2000);

#display the information of books
book1.get_info();
book2.get_info();
```

# **Question#2**



```python
#create a class of person
class Person:
  #attributes of the person
  def __init__(self,name,age):
    self.name=name
    self.age=age

  #property of the person --method that display the name and age of the person
  def get_info(self):
    print(f"The name of the person is {self.name} and its age is {self.age}")


#create a class of student
class Student(Person):
  #attributes of the student
  def __init__(self,name,age,student_id):
    super().__init__(name,age)
    self.student_id=student_id

    #property of the student --method that display the name,age and student_id of the student
  def get_info(self):
    super().get_info()
    print(f"Student ID={self.student_id}\n")

#make the object of the student1
student1=Student("Ayesha",24,1001);
#call the student function
student1.get_info();

student2=Student("Um-e-Hanum",26,1002);
student2.get_info();
```

# **Question#4**

```python
#create a class of student record
class StudentRecord:
  #attributes of the student
  def __init__(self):
    self.studentRecord=[];

 #methods to add student name and score in the studentRecord list

  def add_student(self,name,score):
    self.name=input("Enter student name: ")
    self.score=int(input("Enter student score: "))
    self.studentRecord.append((self.name,self.score))
    print("*******************************")


  #Save the record to a text file records.txt.
  def save_record(self):
    with open('records.txt', 'w') as file:
      for name,score in self.studentRecord:
        file.write(f"Name of Student: {name}, \n Score: {score}\n\n")


  #Read and print all student records from the file.
  def read_record(self):
    with open('records.txt', 'r') as file:
      lines = file.readlines()
      for line in lines:
        print(line.strip())


#Create an instance of StudentRecord and use it to add at least three students. Then, display all the records from the file.
student=StudentRecord();
student.add_student("Ayesha",90);
student.add_student("Um-e-Hanum",80);
student.add_student("Haneeen",70);
student.save_record();
student.read_record();

```
